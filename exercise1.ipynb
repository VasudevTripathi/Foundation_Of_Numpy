unfus(universal functions):

```python
arr=np.array([1,4,9,16])

print(np.sqrt(arr))  #square root
print(np.log(arr))   #natural logarithm
print(np.exp(arr))   #exponential
print(np.sin(arr))   #sine
print(np.cos(arr))   #cosine
print(np.mean(arr))  #mean
print(np.median(arr))#median
print(np.std(arr))   #standard deviation
print(np.sum(arr))   #sum of elements
print(np.min(arr))   #minimum element
print(np.max(arr))   #maximum element
print(np.sort(arr))  #sorted array
```

where in numpy:


```python
arr= np.array([10,-5,20,-3,15])
```

```python
result = np.where(arr>0, 'Positive', 'Negative')
print(result)
```

```python
result= np.where(arr<0,0,arr)  #replace negative values with 0
print(result)
```

```python
arr= np.array([1,2,3,4,5])
```

```python
result= np.where(arr%2 !=0,-1,arr)  #replace odd values with -1
print(result)
```

```python
marks= np.array([45,78,32,90,55])
```

```python
status= np.where(marks>=50,'Pass','Fail')
print(status)
```

```python
data= np.array([10,20,30,40,50])
```

```python
mean_val = np.mean(data)
result= np.where(data> mean_val,mean_val,data)  #replace values greater than mean with mean value
print(result)
```

applying (where) condition in 2d array:

```python
mat= np.array([
    [10,60,30],
    [80,20,90],     
])
```

```python
result= np.where(mat>50,1,0)  #replace values greater than 50 with 1 and others with 0
print(result)
```

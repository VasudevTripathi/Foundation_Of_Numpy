```python
import numpy as np
```

```python
marks = np.array([[78,85,90],[72,68,33],[99,24,56],[88,92,80],[65,84,69]])
```

```python
print(marks.shape)
print(marks.ndim)
print(marks.size)
print(marks.dtype)
```

```python
print(marks[1:2])
print(marks[:,2])
print(marks[0,1])
```

```python
print(np.min(marks))
print(np.average(marks, axis=0))
print(np.average(marks, axis=1))
print(np.max(marks))
```

```python
# identify and display students having avaerage marks >=80
high_achievers= marks[np.average(marks, axis=1) >= 80]
print(high_achievers)
```

```python
# add 5 grace marks to every subject for all students
marks_with_grace= marks + 5
```

```python
#create a status matrix where marks >= 50 is pass(1) and marks < 50 is fail(0)
status_matrix = np.where(marks >= 50, 1, 0)
print(status_matrix)
```
